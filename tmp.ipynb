```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# load dataset
dataframe = pandas.read_excel('task1.xlsx', sheetname='TRAIN')
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,1:16]
Y = dataset[:,16]
```

```python
data = pandas.read_excel('task1.xlsx', sheetname='TRAIN')
tmp = numpy.delete(numpy.asarray(data[data.columns]), [0, 16], 1)

vin1 = numpy.unique(data['VIN_1'])
vin2 = numpy.unique(data['VIN_2'])
vin3 = numpy.unique(data['VIN_3'])
fuelType = numpy.unique(data['FUEL_TYPE'])
bodyType = numpy.unique(data['BODY_TYPE'])
typeOfDrive = numpy.unique(data['TYPE_OF_DRIVE'])
interiorType = numpy.unique(data['INTERIOR_TYPE'])
transmType = numpy.unique(data['TRANSM_TYPE'])
autoCondition = numpy.unique(data['AUTO_CONDITION'])

X = []
for i in tmp:
    arr = []
    for (idx, j) in enumerate(i):
        if idx == 0:
            if numpy.isnan(j):
                arr.append(20)
            else:
                arr.append(j)
        elif idx == 4 or idx == 5 or idx == 6:
            if numpy.isnan(j):
                arr.append(10)
            else:
                arr.append(j)
        elif idx == 7:
            arr.append(j)
        elif idx == 14:
            arr.append(j)
        elif idx == 1:
            k = list(vin1).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 2:
            k = list(vin2).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 3:
            k = list(vin3).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 8:
            k = list(fuelType).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 9:
            k = list(bodyType).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 10:
            k = list(typeOfDrive).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 11:
            k = list(interiorType).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 12:
            k = list(transmType).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
        elif idx == 13:
            k = list(autoCondition).index(j)
            if numpy.isnan(k):
                k = 0
            arr.append(k)
    X.append(arr)
print(X)
```

```python
tmp = numpy.asarray(data[data.columns[16]])
# print(tmp)

Y = []
for i in tmp:
    Y.append(float(i))
print(Y)
```

```python
# define base model
def baseline_model():
	# create model
	model = Sequential()
	model.add(Dense(15, input_dim=15, kernel_initializer='normal', activation='relu'))
	model.add(Dense(1, kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam')
	return model
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn=baseline_model, nb_epoch=100, batch_size=5, verbose=0)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
# evaluate model with standardized dataset
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
# define the model
def larger_model():
	# create model
	model = Sequential()
	model.add(Dense(15, input_dim=15, kernel_initializer='normal', activation='relu'))
	model.add(Dense(6, kernel_initializer='normal', activation='relu'))
	model.add(Dense(1, kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam')
	return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=larger_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Larger: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
# define wider model
def wider_model():
	# create model
	model = Sequential()
	model.add(Dense(20, input_dim=15, kernel_initializer='normal', activation='relu'))
	model.add(Dense(1, kernel_initializer='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam')
	return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=wider_model, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Wider: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
