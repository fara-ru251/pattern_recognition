```python
import numpy as np
vin1 = np.unique(data['VIN_1'])
vin2 = np.unique(data['VIN_2'])
vin3 = np.unique(data['VIN_3'])
fuelType = np.unique(data['FUEL_TYPE'])
bodyType = np.unique(data['BODY_TYPE'])
typeOfDrive = np.unique(data['TYPE_OF_DRIVE'])
interiorType = np.unique(data['INTERIOR_TYPE'])
transmType = np.unique(data['TRANSM_TYPE'])
autoCondition = np.unique(data['AUTO_CONDITION'])
```

```python
tmp = np.delete(np.asarray(data[data.columns]), [0, 16], 1)
# print(tmp)

X = []
for i in tmp:
    arr = []
    for (idx, j) in enumerate(i):
        if idx == 0:
            if np.isnan(j):
                arr.append(1.0)
            else:
                arr.append(float(j) / 20.0)
        elif idx == 4 or idx == 5 or idx == 6:
            if np.isnan(j):
                arr.append(1.0)
            else:
                arr.append(float(j) / 10.0)
        elif idx == 7:
            if np.isnan(j):
                arr.append(1.0)
            else:
                arr.append((float(j) - 1.0) / 5.0)
        elif idx == 14:
            if np.isnan(j):
                arr.append(1.0)
            else:
                arr.append((float(j) - 600000.0) / 22400000.0)
        elif idx == 1:
            k = list(vin1).index(j)
            if np.isnan(k):
                k = len(vin1)
            arr.append(float(k) / float(len(vin1)))
        elif idx == 2:
            k = list(vin2).index(j)
            if np.isnan(k):
                k = len(vin2)
            arr.append(float(k) / float(len(vin2)))
        elif idx == 3:
            k = list(vin3).index(j)
            if np.isnan(k):
                k = len(vin3)
            arr.append(float(k) / float(len(vin3)))
        elif idx == 8:
            k = list(fuelType).index(j)
            if np.isnan(k):
                k = len(fuelType)
            arr.append(float(k) / float(len(fuelType)))
        elif idx == 9:
            k = list(bodyType).index(j)
            if np.isnan(k):
                k = len(bodyType)
            arr.append(float(k) / float(len(bodyType)))
        elif idx == 10:
            k = list(typeOfDrive).index(j)
            if np.isnan(k):
                k = len(typeOfDrive)
            arr.append(float(k) / float(len(typeOfDrive)))
        elif idx == 11:
            k = list(interiorType).index(j)
            if np.isnan(k):
                k = len(interiorType)
            arr.append(float(k) / float(len(interiorType)))
        elif idx == 12:
            k = list(transmType).index(j)
            if np.isnan(k):
                k = len(transmType)
            arr.append(float(k) / float(len(transmType)))
        elif idx == 13:
            k = list(autoCondition).index(j)
            if np.isnan(k):
                k = len(autoCondition)
            arr.append(float(k) / float(len(autoCondition)))
    X.append(arr)
print(X)
```

```python
tmp = np.asarray(data[data.columns[16]])
# print(tmp)

Y = []
for i in tmp:
    Y.append((float(i) - 500000.0) / 24500000.0)
print(Y)
```

```python
# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
import matplotlib

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
# dataset = numpy.loadtxt("iris.txt", delimiter=",")
#dataset = numpy.loadtxt("pima-indians-diabetes.data", delimiter=",")
# split into input (X) and output (Y) variables
# X = dataset[:,0:4]
# Y = dataset[:,4:7]
# create model
model = Sequential()
model.add(Dense(32, input_dim=15, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=50, batch_size=10,  verbose=2)
# calculate predictions
predictions = model.predict(X)
```

```python
print(predictions)
```

```python
rounded = [float((x*24500000.0)+500000.0) for x in predictions]
print(rounded)
```

```python
result = [float((x*24500000.0)+500000.0) for x in Y]
c = 0
ttl = 0
for (idx, val) in enumerate(result):
    k = abs(rounded[idx]-val)
    if k < val * 0.1:
        c += 1
    ttl += ((k * 100) / val)
print(ttl / len(result))
print (c * 100) / 6000, "%"
```
