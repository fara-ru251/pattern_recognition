```python
import numpy as np
vin1 = np.unique(data['VIN_1'])
vin2 = np.unique(data['VIN_2'])
vin3 = np.unique(data['VIN_3'])
fuelType = np.unique(data['FUEL_TYPE'])
bodyType = np.unique(data['BODY_TYPE'])
typeOfDrive = np.unique(data['TYPE_OF_DRIVE'])
interiorType = np.unique(data['INTERIOR_TYPE'])
transmType = np.unique(data['TRANSM_TYPE'])
autoCondition = np.unique(data['AUTO_CONDITION'])
```

```python
tmp = np.delete(np.asarray(data[data.columns]), [0, 16], 1)
# print(tmp)

X = []
for i in tmp:
    arr = []
    for (idx, j) in enumerate(i):
        if idx == 0:
            arr.append(float(j) / 19.0)
        elif idx == 4 or idx == 5 or idx == 6:
            arr.append(float(j) / 9.0)
        elif idx == 7:
            arr.append((float(j) - 1.0) / 4.6)
        elif idx == 14:
            arr.append((float(j) - 600000.0) / 22400000.0)
        elif idx == 1:
            k = list(vin1).index(j)
            arr.append(float(k) / float(len(vin1) - 1))
        elif idx == 2:
            k = list(vin2).index(j)
            arr.append(float(k) / float(len(vin2) - 1))
        elif idx == 3:
            k = list(vin3).index(j)
            arr.append(float(k) / float(len(vin3) - 1))
        elif idx == 8:
            k = list(fuelType).index(j)
            arr.append(float(k) / float(len(fuelType) - 1))
        elif idx == 9:
            k = list(bodyType).index(j)
            arr.append(float(k) / float(len(bodyType) - 1))
        elif idx == 10:
            k = list(typeOfDrive).index(j)
            arr.append(float(k) / float(len(typeOfDrive) - 1))
        elif idx == 11:
            k = list(interiorType).index(j)
            arr.append(float(k) / float(len(interiorType) - 1))
        elif idx == 12:
            k = list(transmType).index(j)
            arr.append(float(k) / float(len(transmType) - 1))
        elif idx == 13:
            k = list(autoCondition).index(j)
            arr.append(float(k) / float(len(autoCondition) - 1))
    X.append(arr)
print(X)
```

```python
tmp = np.asarray(data[data.columns[16]])
# print(tmp)

Y = []
for i in tmp:
    Y.append((float(i) - 500000.0) / 24500000.0)
print(Y)
```

```python
# Create first network with Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
import matplotlib

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# load pima indians dataset
# dataset = numpy.loadtxt("iris.txt", delimiter=",")
#dataset = numpy.loadtxt("pima-indians-diabetes.data", delimiter=",")
# split into input (X) and output (Y) variables
# X = dataset[:,0:4]
# Y = dataset[:,4:7]
# create model
model = Sequential()
model.add(Dense(12, input_dim=15, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=10, batch_size=10,  verbose=2)
# calculate predictions
predictions = model.predict(X)
```

```python
print(predictions)
```

```python
rounded = [round(x) for x in predictions]
print(rounded)
```

```python
d = [[0, 0], [0, 0]]
def check(index, x):
    ans = (Y[index] == x)
    if x == 0.0:
        if ans:
            d[0][0] += 1;
        else:
            d[0][1] += 1;
    else:
        if ans:
            d[1][1] += 1;
        else:
            d[1][0] += 1;
    return 
result = [check(index, x) for index, x in enumerate(rounded)]

print(d)
```

```python
from matplotlib import pyplot as plt
import numpy as np

# vals = np.array([[1, 2, 3, 4], [2, 3, 4, 5], [3, 4, 5, 6]])
vals = np.array(d)
fig, ax = plt.subplots()
ax.pie(vals.flatten(), radius=1.2,
       colors=plt.rcParams["axes.prop_cycle"].by_key()["color"][:vals.shape[1]])
ax.pie(vals.sum(axis=1), radius=1)
ax.set(aspect="equal", title='Pie plot with `ax.pie`')

plt.show()
```
